```python
# loading dataset
url = '/content/Telcom_dataset.csv'
df_tel = pd.read_csv(url)
df_tel.head(5)
```

```python
# loading dataset
url = '/content/Telcom_dataset2.csv'
df_tel2 = pd.read_csv(url)
df_tel2.head(5)
```

```python
# loading dataset
url = '/content/Telcom_dataset3.csv'
df_tel3 = pd.read_csv(url)
df_tel3.head(5)
```

```python
# merging
result1 = pd.merge(df_tel, df_tel2, left_index=True, right_index=True, how="outer")
result1
```

```python
# merging all telcom data
merged_telcom = pd.merge(result1, df_tel3, left_index=True, right_index=True, how="outer")
merged_telcom
```

```python
# checking the last details
df_geo.tail(5)
```

```python
# merging ville with the telcom data
result5 = pd.merge(df_geo['VILLES'], merged_telcom, left_index=True, right_index=True, how="outer")
result5
```

```python
# 1) Which ones were the most used city for the last three days

filtered_df = result5.loc[(result5['DATETIME'] >= '2012-05-07')]
filtered_df
filtered_df['VILLES'].value_counts()
```
